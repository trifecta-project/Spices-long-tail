```python
import pandas as pd
import matplotlib.pyplot as plt

spice_data = pd.read_csv('Spice_data_clean.csv')

peper_data = spice_data[spice_data['Cargo'] == 'peper']
kaneel_data = spice_data[spice_data['Cargo'] == 'kaneel']

voyages_per_year_peper = peper_data.groupby('Book Year')['Voyage'].count()
voyages_per_year_kaneel = kaneel_data.groupby('Book Year')['Voyage'].count()

plt.figure(figsize=(14, 7))
plt.plot(voyages_per_year_peper.index, voyages_per_year_peper, marker='o', color='orange', linewidth=2)
plt.title('Number of Voyages for Peper Over Time', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Voyages', fontsize=14)
plt.grid(axis='both', linestyle='--', alpha=0.7)

plt.xticks(range(min(voyages_per_year_peper.index), max(voyages_per_year_peper.index)+1, 5))

plt.tight_layout()
plt.savefig('voyages_peper.png')
plt.show()

plt.figure(figsize=(14, 7))
plt.plot(voyages_per_year_kaneel.index, voyages_per_year_kaneel, marker='o', color='purple', linewidth=2)
plt.title('Number of Voyages for Kaneel Over Time', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Voyages', fontsize=14)
plt.grid(axis='both', linestyle='--', alpha=0.7)

plt.xticks(range(min(voyages_per_year_kaneel.index), max(voyages_per_year_kaneel.index)+1, 5))

plt.tight_layout()
plt.savefig('voyages_kaneel.png')
plt.show()
```
